# Data Preprocessing

Selecting the columns that to keep in analysis:-

1) The 'index' column is a duplicate so it has to drop.


2) The 'Job Description' won't be accurate neither as each row contains a high amount of text therefore impossible to convert in numeric values.

3) The 'Competitors' column contains mainly '-1' values so we won't be able to analyze it.

```python
df.columns
```

```python
df.drop(['Competitors','Job Description','index'], axis=1, inplace=True)
```

Now we want to remove every rows containing the '-1' values in the 'Headquarters', 'Founded' and 'Industry' columns:-

```python
df_new=df.drop(df[(df.Headquarters == '-1') | (df.Founded == -1) | (df.Industry == -1) | (df.Sector == '-1') | (df.Rating == -1)].index)
```

Then we remove the duplicates

```python
df_new.dropna(inplace=True)
df_new.drop_duplicates(inplace=True)
```

# Steps to Perform
### 1) Salary Estimate

We can notice undesired characters at the end of each value combined with the rating of the company in the 'Company Name'.

Similarly, we don't need the information '(Glassdor est.)' in the 'Salary Estimate' column.

```python
df_new['Company Name']=df_new['Company Name'].str.split('\n').str[0]
df_new['Salary Estimate']=df_new['Salary Estimate'].str.split('(').str[0]
```

```python
df_new
```

The 'Salary Estimate' column can't be analyzed as it is.

Therefore we can split the maximum and minimum salary using the str.split method. 

Also we can do this for 'Headquarters' column as well to have the city and the state separated:

```python
df_new2=df_new
df_new2[['Min_salary_estimate','Max_salary_estimate']] = df_new2['Salary Estimate'].str.split('-',expand=True)
```

```python
df_new2.drop(['Salary Estimate'],axis=1,inplace=True)
df_new2
```

Both columns still contain undesired string values. We can use the str.replace method to get rid of those:

```python
df_new2['Min_salary_estimate'] = df_new2['Min_salary_estimate'].str.replace(r'\D',"")
df_new2['Max_salary_estimate'] = df_new2['Max_salary_estimate'].str.replace(r'\D',"")
```

And convert it to numeric values

```python
df_new2['Min_salary_estimate']=df_new2['Min_salary_estimate'].astype(np.int64)
df_new2['Max_salary_estimate']=df_new2['Max_salary_estimate'].astype(np.int64)
```

```python
df_new2['Min_salary_estimate'] = df_new2['Min_salary_estimate'].multiply(1000)
df_new2['Max_salary_estimate'] = df_new2['Max_salary_estimate'].multiply(1000)
df_new2
```

### 2) Headquarters and Location

Regarding the 'Headquarters' and 'Location' columns we want to have the city and state in a separate location:

```python
df_new2[['HQ_city','HQ_state']]=df_new2['Headquarters'].str.split(',',expand=True)
df_new2['Job_city'] = df_new2.Location.str.split(',', expand = True)[0]
df_new2['Job_state'] = df_new2.Location.str.split(',', expand = True)[1]
df_new2.drop(['Headquarters','Location'], axis=1, inplace=True)
df_new2
```

### 3) Size

```python
df_new2['Size']=df_new2['Size'].str.replace(' employees','')
df_new2['Size']=df_new2['Size'].str.replace(' to ','-')
df_new2
```

### Founded

```python
today=dt.date.today()
df_new2['Company_age']=today.year-df_new2['Founded']
df_new2
```

### Revenue

The value 'Unknown / Non-Applicable' can be converted into 'N/A' for more readibility:

```python
df_new2['Revenue'].value_counts()
```

```python
df_new2['Revenue'] = df_new2['Revenue'].str.replace('Unknown / Non-Applicable',"N/A")
df_new2['Revenue'] = df_new2['Revenue'].str.replace('USD','')
df_new2['Revenue'] = df_new2['Revenue'].str.replace('[()]','')
df_new2['Revenue'] = df_new2['Revenue'].str.replace('to','-')
df_new2
```

Now let's create a new column named 'Avg_salary_estimate' to get the average salary estimated:

```python
df_new2['Avg_salary_estimate']=df_new2[['Min_salary_estimate','Max_salary_estimate']].mean(axis=1)
df_new2['Avg_salary_estimate']=df_new2['Avg_salary_estimate'].astype('int')
```

Now we notice the information 'Senior' in certain rows on the 'Job Title' column. 

We can therefore create a new column 'Seniority' to identify which job offer requires to be Senior or Junior:

```python
df_new2['Seniority'] = df_new2['Job Title'].str.contains('Senior',case=False)
df_new2['Seniority'] = df_new2['Seniority'].astype(str)
df_new2['Seniority'] = df_new2['Seniority'].str.replace('False','Junior')
df_new2['Seniority'] = df_new2['Seniority'].str.replace('True','Senior')
```

```python
df_new2['Seniority'].value_counts()
```

```python
df_finaldata=df_new2
df_finaldata
```

# Plotting Graphs 

### 1) Number of job offers per Seniority

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.countplot(x='Seniority', data=df_finaldata)
plt.xlabel("Job seniority", fontsize=12)
plt.ylabel("Number of job offers", fontsize=12)
plt.title("Number of job offers per Seniority", fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.show()
```

### 2) Number of job offers per Seniority and Sector

```python
figure(figsize=(12, 8), dpi=80)
sns.countplot(data=df_finaldata, x=df_finaldata['Sector'],hue=df_finaldata['Seniority'], order=df_finaldata['Sector'].value_counts().index)
plt.xticks(rotation=45,ha='right', va='top', fontsize=11)
plt.yticks(fontsize=11)
plt.xlabel("Sector",fontsize=12)
plt.ylabel("Number of job offers",fontsize=12)
plt.title("Number of job offers per Seniority and Sector",fontsize=15)
plt.show()
```

### 3) Average estimate salary per Sector

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.barplot(x='Sector', y='Avg_salary_estimate', data=df_finaldata, order=df_finaldata.groupby(["Sector"])['Avg_salary_estimate'].mean().sort_values().index,ci=None)
plt.xlabel("Sector", fontsize=12)
plt.ylabel("Average estimate salary", fontsize=12)
plt.title("Average estimate salary per Sector", fontsize=15)
plt.xticks(rotation=30, ha='right', va='top', fontsize=11)
plt.yticks(fontsize=11)
plt.show()
```

### 4) Correlation between the average salary and the company rating

```python
figure(figsize=(12, 8), dpi=80)
sns.scatterplot(data=df_finaldata, x='Avg_salary_estimate', y='Rating', hue='Seniority')
plt.xlabel("Average estimate salary", fontsize=12)
plt.ylabel("Rating", fontsize=12)
plt.title("Correlation between the average salary and the company rating", fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(title='Seniority', fontsize=11)
plt.show()
```

### 5) Distribution of the average estimate salary per type of ownership

```python
figure(figsize=(12, 8), dpi=80)
sns.histplot(df_finaldata,x=df_finaldata['Avg_salary_estimate'], bins=15, hue=df_finaldata['Type of ownership'])
plt.xlabel("Average estimate salary",fontsize=12)
plt.ylabel("Number of job offers",fontsize=12)
plt.title("Distribution of the average estimate salary per type of ownership",fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.show()
```

### 6) Distribution of the average salary in each company size

```python
figure(figsize=(12, 8), dpi=80)
sns.boxplot(x=df_finaldata['Size'], y=df_finaldata['Avg_salary_estimate'])
plt.xlabel("Company size", fontsize=12)
plt.ylabel("Average estimate salary", fontsize=12)
plt.title("Distribution of the average salary in each company size", fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.show()
```

### 7) Number of job offers per Company size

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.countplot(x=df_finaldata['Size'])
plt.xlabel("Company size", fontsize=12)
plt.ylabel("Number of job offers", fontsize=12)
plt.title("Number of job offers per Company size", fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.show()
```
